Markdown Commentary: What, Why, and Results

```python
# Q5 Linear Regression Model 1: Taxes as Predictor

# Define the dependent variable (Y) and independent variable (X)
Y = df_cleaned['Sell']
X1 = df_cleaned['Taxes']

# Add a constant  for the intercept (b0)
X1 = sm.add_constant(X1)

# Fit the OLS model
model_taxes = sm.OLS(Y, X1).fit()

# Display the model output
print("--- Linear Regression Model: Sell vs. Taxes ---")
print(model_taxes.summary())
```

# Q6. Markdown Commentary on Model 1 Output

## Is Taxes a good predictor for selling price? Why or why not?

Yes, Taxes is a good, statistically significant predictor for selling price.
Taxes is a strong predictor of selling price, showing a clear positive and statistically significant relationship (R² = 0.81, p < 0.001).
In simple terms, homes with higher property taxes generally sell for more, although taxes alone don’t explain every difference in price.

```python
# Q7 Linear Regression Model 2: Living Area as Predictor
# Define the new independent variable (X)
X2 = df_cleaned['Living']

# Add a constant term for the intercept (b0)
X2 = sm.add_constant(X2)

# Fit the OLS model
model_living = sm.OLS(Y, X2).fit()

# Display the model output
print("--- Linear Regression Model: Sell vs. Living Area ---")
print(model_living.summary())
```

# Q7. Which Model is Better?

### Markdown Commentary on Model Comparison and Limitations

This analysis explored how different housing features relate to selling price using **50** observations and **9** key variables.
The results show that List Price, Taxes, and Living Area all have strong positive relationships with selling price.
The Sell vs. List Price model achieved an R² of 0.995, meaning list price explains nearly all of the variation in selling price — though this is expected since homes are typically priced close to their market value.
The Sell vs. Taxes model showed an **R² of 0.81**, indicating that homes with higher property taxes tend to sell for more, while the Sell vs. Living Area model had an  **R² of 0.77**, showing that larger homes generally command higher prices.

In simple terms, homes that are larger, more highly taxed, and listed at higher prices tend to sell for more — but each factor alone doesn’t tell the full story.
A multiple regression model combining several predictors would likely provide a more complete and accurate view of what drives home values.

The List Price model fits the data best (R² = 0.995), explaining almost all variation in selling price.
However, Taxes is a more independent and realistic predictor of home value, even though it explains slightly less (R² = 0.81).
