```python
#11 part a (statistics of wage)
print("Mean wage is :",df.wage.mean())
print("Std Deviation of wages are :",df.wage.std())
print("Median wage is ",df.wage.median())
print("Minimum wage is",df.wage.min())
print("Maximum wage is ",df.wage.max())
```

```python
#11 part b (statistics of age)
print("Mean age is :",df.age.mean())
print("Std Deviation of ages are :",df.age.std())
print("Median age is ",df.age.median())
print("Minimum age is",df.age.min())
print("Maximum age is ",df.age.max())
```

```python
#12 part a (statistics of wage by race)
print("Mean wage is :",df.groupby('race')['wage'].mean())
print("Std Deviation of wages are :",df.groupby('race')['wage'].std())
print("Median wage is ",df.groupby('race')['wage'].median())
print("Minimum wage is",df.groupby('race')['wage'].min())
print("Maximum wage is ",df.groupby('race')['wage'].max())
```

```python
#12 part b (statistics of age by race)
print("Mean age is :",df.groupby('race')['age'].mean())
print("Std Deviation of ages are :",df.groupby('race')['age'].std())
print("Median age is ",df.groupby('race')['age'].median())
print("Minimum age is",df.groupby('race')['age'].min())
print("Maximum age is ",df.groupby('race')['age'].max())
```

```python
#18
y=df['race'].value_counts(normalize=True) * 100
```

```python
#19
import matplotlib.pyplot as plt

y.plot(kind='bar', color='skyblue')

plt.title("Proportion of Values in race")
plt.xlabel("Values")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=45)  # rotate x-labels for readability
plt.show()
```
