#### 6. Use Python to determine how many days until New Year's Day

```python
import datetime

def days_to_new_year():
    """
    Calculates and prints the number of days until New Year's Day.
    """
    # Get today's date
    today = datetime.date.today()

    # Determine the next New Year's Day
    new_year = datetime.date(today.year + 1, 1, 1)

    # Calculate the difference between the two dates
    diff = new_year - today

    # Print the result
    print(f"There are {diff.days} days until New Year's Day.")

# Call the function
days_to_new_year()
```

#### 7. Write a short program that displays the future value of 1,000 earning 5%  at the end of each year for the next 5 years

```python
# Initial investment parameters
principal = 1000
rate = 0.05
years = 5
for year in range(1, years + 1):
  
    principal = principal * (1 + rate)

   
    print(f"End of Year {year}: ${principal:,.2f}")
```

#### 8. Write a short program that prompts a user to enter a stock symbol and press enter. Add each symbol entered to a variable. The program should run until the user presses the enter key without entering anything (empty string) 

```python
# Create an empty list to store the stock symbols
stock_symbols = []

while True:
    # Prompt the user for a stock symbol
    symbol = input("Enter a stock symbol (or press Enter to finish): ")

    # Check if the user entered an empty string
    if symbol == "":
        break  # Exit the loop if the input is empty
    else:
        # Add the entered symbol to the list
        stock_symbols.append(symbol)

# Print the final list of stock symbols
print("\nStock symbols entered:", stock_symbols)
```

#### 9. Write the code necessary to calculate the sum of the square differences of a group of values : 
58, 32, 37, 41, 36, 36, 54, 37, 25, 53
##### $\Sigma $(${x}$ - $\bar{x}$)$^2$

```python
import numpy as np

# The list of values
values = [58, 32, 37, 41, 36, 36, 54, 37, 25, 53]

# Calculate the mean of the values
mean = np.mean(values)

# Calculate the sum of the square differences from the mean
sum_sqr_diff = np.sum((values - mean)**2)

# Print the final result
print(f"The values are: {values}")
print(f"The mean of the values is: {mean}")
print(f"The sum of the square differences is: {sum_sqr_diff}")
```

#### 10. Opening, reading and writing files

```python
def write_and_read_file(filename, content):
    """
    Writes content to a file and then reads it back.

    Args:
        filename (str): The name of the file to create and use.
        content (str): The text content to write to the file.
    """
    # --- Writing to a file ---
    print(f"--- Writing content to '{filename}' ---")
    try:
     
        with open(filename, 'w') as file:
            file.write(content)
        print("Successfully wrote content to the file.")
    except IOError as e:
        print(f"An error occurred while writing to the file: {e}")
        return  # Exit the function if writing fails

    print("\n--- Reading content from the file ---")
    # --- Reading from a file ---
    try:
        # Open the file in 'read' mode ('r'). This is the default mode.
        with open(filename, 'r') as file:
            # Read all the content from the file into a variable.
            read_content = file.read()
            print("Successfully read content from the file.")
            print("--- File Content ---")
            print(read_content)
            print("--------------------")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except IOError as e:
        print(f"An error occurred while reading the file: {e}")

if __name__ == "__main__":
    # Define the name of the file to be created
    my_filename = "my_text_file.txt"

    # Define the content to be written to the file
    file_content = """This is the first line of my text file.
This is the second line.
And this is the third and final line.
"""
    # Call the function to perform the operations
    write_and_read_file(my_filename, file_content)
```

#### 11. Create a function the will simulate rolling two die and return the value of each and and the sum 

```python
import random

def roll_two_dice():
    """
    Simulates the rolling of two six-sided dice.

    Returns:
        A dictionary containing the value of each die and their sum.
        Example: {'die1': 4, 'die2': 6, 'sum': 10}
    """

    die1 = random.randint(1, 6)
    
    
    die2 = random.randint(1, 6)
    

    total_sum = die1 + die2

    return {
        'die1': die1,
        'die2': die2,
        'sum': total_sum
    }

if __name__ == "__main__":
    results = roll_two_dice()
    print(f"Rolling the dice...")
    print(f"First die: {results['die1']}")
    print(f"Second die: {results['die2']}")
    print(f"Total sum: {results['sum']}")
```

#### 12. Write the code needed to evaluate whether a value in a group of values is negative or postive and prints an approriate message, i.e. postive, negative 
-1.59, 2.36, 1.69, 1.13, -0.91, 1.48, -0.34, 1.31, -0.74, 0.2

```python
# A group of values to evaluate
values = [10, -5, 0, 25, -100, 3.14, -0.01]

# Loop through each value in the group
for value in values:
    # Check if the value is positive, negative, or zero
    if value > 0:
        print(f"The value {value} is positive.")
    elif value < 0:
        print(f"The value {value} is negative.")
    else:
        print(f"The value {value} is zero.")
```

#### 14.  Complete the following:
1. Write a program that generates 500 random integers from 1 to n, for example, the first integer will be 1 the second will be 1 or 2, the fifth between 1 and 5, and so on.
2. The program should write each integer to a file 
3. Once all integers are written close the file

##### Part 2
1. Open and read the file into a list
2. Create a line plot of the data

```python
import random

# Number of integers to generate
num_integers = 500
file_name = "random_numbers.txt"

# Open the file in 'write' mode ('w').
# The 'with' statement ensures the file is automatically closed
# even if an error occurs.
with open(file_name, 'w') as file:
    # Loop from 1 up to (and including) num_integers
    for i in range(1, num_integers + 1):
        # Generate a random integer from 1 to i
        random_number = random.randint(1, i)
        
        # Write the number to the file.
        # We convert it to a string and add a newline character
        # so each number is on a separate line.
        file.write(str(random_number) + '\n')

print(f"Successfully generated and wrote {num_integers} integers to '{file_name}'.")
```

```python
import matplotlib.pyplot as plt

file_name = "random_numbers.txt"
data = []

# Open the file in 'read' mode ('r')
try:
    with open(file_name, 'r') as file:
        # Read each line from the file
        for line in file:
            # Convert the string to an integer and add it to the list
            data.append(int(line.strip()))
except FileNotFoundError:
    print(f"Error: The file '{file_name}' was not found. Please run Part 1 first.")
    exit()

# Create the x-axis values (from 1 to the number of data points)
x_values = list(range(1, len(data) + 1))

# Create the plot
plt.figure(figsize=(10, 6))  # Optional: Adjust the size of the plot
plt.plot(x_values, data)

# Add a title and labels for clarity
plt.title("Random Integers (1 to n) Over 500 Trials")
plt.xlabel("Trial Number (n)")
plt.ylabel("Generated Random Integer")
plt.grid(True)
plt.tight_layout()  # Adjusts plot to prevent labels from being cut off

# Display the plot
plt.show()
```
